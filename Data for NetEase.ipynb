```python
import wrds
import pandas as pd

db = wrds.Connection()

# 只查近10年（假设今天是2026年，就查2016年之后）
query = """
select tic, datadate, revtq, niq
from comp.fundq
where tic='NTES' 
  and datadate >= '2016-01-01'
order by datadate desc
"""
data = db.raw_sql(query)
```

```python
print(data)
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# 把 data 改成 df 的名字，或者直接用 data
df = data  # 你之前存数据的变量叫 data

# 数据清洗
df['datadate'] = pd.to_datetime(df['datadate'])
df['revtq'] = pd.to_numeric(df['revtq'], errors='coerce')
df = df.dropna(subset=['revtq'])
df = df.sort_values('datadate').reset_index(drop=True)

# 加季度序号
df['quarter_num'] = range(1, len(df) + 1)

# 训练
X = df[['quarter_num']]
y = df['revtq']
model = LinearRegression()
model.fit(X, y)

# 预测
last_q = df['quarter_num'].max()
future_q = [[last_q + i] for i in range(1, 5)]
predictions = model.predict(future_q)

print("预测结果：")
for i, pred in enumerate(predictions, 1):
    print(f"第{i}季度: ${pred:.0f}M")
```

```python
data.to_csv('netease_revenue.csv', index=False)
```
